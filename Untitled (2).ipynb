```python
import hashlib

def generate_short_hash(Customer_Name):
    # Calculate the MD5 hash of the customer name
    hashed = hashlib.md5(Customer_Name.encode()).hexdigest()
    # Truncate the hash to 6 characters
    short_hash = hashed[:6]
    return short_hash

# Assuming df is your DataFrame and 'Customer_Name' is the column containing customer names
df['Cust_ID'] = df['Customer_Name'].apply(generate_short_hash)

# Displaying the DataFrame with the generated short IDs
df[['Customer_Name', 'Cust_ID']]
```

# Outlier Analysis

```python
#Monetary
Q1 = merge_cols.Monetary.quantile(0.05)
Q3 = merge_cols.Monetary.quantile(0.95)
IQR = Q3 - Q1
merge_cols = merge_cols[(merge_cols.Monetary >= Q1 - 1.5 * IQR) & (merge_cols.Monetary <= Q3 + 1.5 * IQR)]

#Frequency
Q1 = merge_cols.Frequency.quantile(0.05)
Q3 = merge_cols.Frequency.quantile(0.95)
IQR = Q3 - Q1
merge_cols = merge_cols[(merge_cols.Frequency >= Q1 - 1.5 * IQR) & (merge_cols.Frequency <= Q3 + 1.5 * IQR)]



```

# Scaling

```python


# Assuming 'df_scaled' is your DataFrame with Recency, Frequency, and Monetary columns
scale = merge_cols[['Recency', 'Frequency', 'Monetary']]

# Convert 'Recency' column from Timedelta to number of days (float)
scale['Recency'] = scale['Recency'].dt.days

# Initialize the StandardScaler
scaler = StandardScaler()

# Scale the data
df_scaled = scaler.fit_transform(scale)
df_scaled.shape
```

# Model Building

```python

# Create an empty list to store the sum of squared distances (SSDs)
ssd = []

# Define the range of clusters to iterate over
range_n_clusters = [2, 3, 4,5,6,7,8]

# Fit KMeans models for each number of clusters in the range
for num_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=num_clusters, max_iter=50)
    kmeans.fit(df_scaled)
    ssd.append(kmeans.inertia_)
plt.plot(ssd)
```

```python
kmeans= KMeans(n_clusters = 4, init = 'k-means++') #initialize the class object
label= kmeans.fit_predict(df_scaled) #returns a cluster number for each of the data points
print(label)
```

```python
# Convert necessary data types if required
df['Sales'] = df['Sales'].astype(int)  # Assuming 'Sales' needs to be an integer type

# Create a simplified DataFrame of relevant information
item_purchases = df[['Cust_ID', 'Product_Name', 'Customer_Segment', 'Product_Category', 'Product_Sub_Category']].copy()
```

```python
# Compute the frequency of each product being bought within each customer segment and category
product_freq = df.groupby(['Customer_Segment', 'Product_Category', 'Product_Name']).size().reset_index(name='Frequency')

# Filter to get the most popular items per segment and category
popular_products = product_freq.groupby(['Customer_Segment', 'Product_Category']).apply(lambda x: x.nlargest(3, 'Frequency')).reset_index(drop=True)
```

```python
def recommend_products(cust_id, customer_segment, product_category, num_recommendations=3):
    # Identify products the customer has already purchased
    existing_products = df[(df['Cust_ID'] == cust_id) & (df['Product_Category'] == product_category)]['Product_Name'].unique()
    
    # Filter the popular products for the segment and category
    recommendations = popular_products[(popular_products['Customer_Segment'] == customer_segment) & (popular_products['Product_Category'] == product_category)]
    
    # Exclude already purchased products
    recommendations = recommendations[~recommendations['Product_Name'].isin(existing_products)].head(num_recommendations)
    
    return recommendations[['Product_Name', 'Frequency']]


```

```python

# Example: Recommend 3 products for a customer in the 'Corporate' segment interested in 'Office Supplies'
customer_id = '00a6fb'
customer_segment = 'Corporate'
product_category = 'Office Supplies'
recommendations = recommend_products(customer_id, customer_segment, product_category)
```
